# Problem 2

An important part of programming is learning to interpret error messages and understanding what correction needs to be made.

Read and familiarize yourself with the following error messages.

Explain the error. Then duplicate each cell and correct the error. The first problem has been done for you as an example.

```python
# A
print("Hello World"
```

Answer: The `print()` function is missing the closing parenthesis. This results in an unexpected EOF error.

```python
# corrected:
print("Hello World")
```

```python
# B
print("Hello")
    print("Goodbye")
```

Answer: The second `print()` function should not be indented. This results in an unexpected indentation error. In Python, consistent indentation is used to define the structure of code blocks.

```python
# corrected:
print("Hello")
print("Goodbye")
```

```python
# C
x = 10
if x > 8
    print("x is greater than 8")
```

Answer: The `if` statement should have a colon `:` after the condition. In Python, every if statement must be followed by a colon to start the block of code that belongs to the `if` condition. This results in a syntax error.

```python
# corrected
x = 10
if x > 8:
    print("x is greater than 8")
```

```python
# D
if x = 10:
    print("x is equal to 10")
```

Answer: The condition in the `if` statement should have a conditional operator like `==` or `:=`. `=` is used to assign values to variables. This results in a syntax error.

```python
# corrected:
if x == 10:
    print("x is equal to 10")
```

```python
# E
x = 5
if x == 5:
print("x is five")
```

Answer: The `print` statement should be indented after the `if` statement to indicate that the line should run if the `if` statement is true. This results in an indentation error.

```python
# corrected:
x = 5
if x == 5:
    print("x is five")
```

```python
# F
l = [1, 2, 50, 10]
l = sort(l)
```

Answer: The `sort` function is modifying l directly without outputting any value. Therefore, assigning it back to l does not make sense as there is no output value. This results in a name error.

```python
# corrected:
l = [1, 2, 50, 10]
l.sort()
print(l)
```

# Problem 3

Use Python as a calculator. Enter the appropriate calculation in a cell and be sure the output value is visible.

A. How many seconds are there in 42 minutes 42 seconds?

```python
(42 * 60) + 4
```

B. There are 1.61 kilometers in a mile. How many miles are there in 10 kilometers?

```python
10 / 1.61
```

C. If you run a 10 kilometer race in 42 minutes 42 seconds, what is your average 1-mile pace (time to complete 1 mile in minutes and seconds)? What is your average speed in
miles per hour?

```python
second_per_mile = 2562 / (10 / 1.61)
minutes = int(second_per_mile / 60)
seconds = second_per_mile % 60
print(f"It will take {minutes} minutes and {seconds} seconds to complete 1 mile.")

hours = 2562 / 3600
mile_per_hour = (10 / 1.61) / hours
print(f"The average speed is {mile_per_hour} miles per hour.")
```
